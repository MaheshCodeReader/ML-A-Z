```python
## Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Import dataset
dataset = pd.read_csv('Mall_Customers.csv')

## matrix of features
X = dataset.iloc[:, [3,4]].values
# .values converts the spliced dataframe into 2d array
```

```python
# Using the dendrogram to determine number of clusters
#%matplotlib qt
%matplotlib inline
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# As can be seen, 5 is optimal number of clusters
```

```python
# fitting AgglomerativeCLustering to dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
y_hc
```

```python
# Visualizing the clusters
# Visualizing clusters
## conditional indices in python

plt.scatter(X[:,0], X[:, 1])
plt.title('all points')
plt.show()
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0,1], c = 'red', label='cluster 0')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1,1], c = 'green', label='cluster 1')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2,1], c = 'blue', label='cluster 2')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3,1], c = 'yellow', label='cluster 3')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4,1], c = 'black', label='cluster 4')
#plt.scatter(y_hc.cluster_centers_[:, 0], y_hc.cluster_centers_[:, 1], c = 'magenta', label='Centroids', s = 100)
# cluster_centers_ is used in kmeans, and I copy-pasted the code
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.title('After clustering')
plt.show()
```

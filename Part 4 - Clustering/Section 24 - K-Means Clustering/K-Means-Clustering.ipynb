```python
## Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Import dataset
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
# matrix of features
X = dataset.iloc[:, [3,4]].values

## There is no dependent variable in clustering
```

```python
## ELBOW METHOD to determine k
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow Method Graph')
plt.xlabel('number of centroids')
plt.ylabel('inertia WCSS')
plt.show()
```

```python
## It appears k = 5 is a good number
```

```python
# choosing n_clusters=5
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
# Visualizing clusters
## conditional indices in python

plt.scatter(X[:,0], X[:, 1])
plt.title('all points')
plt.show()
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,1], c = 'red', label='cluster 0')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1,1], c = 'green', label='cluster 1')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2,1], c = 'blue', label='cluster 2')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3,1], c = 'yellow', label='cluster 3')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4,1], c = 'black', label='cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = 'magenta', label='Centroids', s = 100)
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.title('After clustering')
plt.show()
```
